```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_openml
df = fetch_openml(name="house_prices")
```

```python
df
```

***Preparing the dataset***

```python
dataset=pd.DataFrame(df.data)
dataset.head()
```

```python
dataset['Price']=df.target
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

***Checking the missing values***

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(dataset.isnull(),cbar=False)
```

```python
dataset = dataset.dropna(axis=1)
```

```python
sns.heatmap(dataset.isnull(),cbar=False)
```

***Removing the string values***

```python
dataset.corr()
```

```python
dataset = dataset.applymap(lambda x: np.nan if isinstance(x, str) else x) #To replace all string data in a DataFrame with np.nan (missing value)
```

```python
sns.heatmap(dataset.isnull(),cbar=False)
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna(axis=1)
```

```python
sns.heatmap(dataset.isnull(),cbar=False)
```

```python
dataset.corr()
```

***Separating the independent and dependent variables***

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
X
```

```python
Y
```

***Train and test split***

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
plt.scatter(dataset['OverallQual'], dataset['Price'])
plt.xlabel('Overall Quality')
plt.ylabel('Price')
```

```python
sns.regplot(x='OverallQual', y='Price', data=dataset)
```

***Standarizing the dataset***

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #values of features in dataset are calculated using different units so we need to make it one standard
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

***Model Training***

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

***Prediction with the test data***

```python
reg_pred=regression.predict(x_test)
```

```python
reg_pred
```

```python
y_test
```

```python
import seaborn as sns

ax1 = sns.distplot(y_test, hist=False, color="r", label="Actual Value")
sns.distplot(reg_pred, hist=False, color="b", label="Fitted Values" , ax=ax1)
```

```python
plt.scatter(y_test, reg_pred)
```

```python
residuals = y_test - reg_pred
residuals
```

```python
sns.displot(residuals, kind = 'kde')
```

```python
plt.scatter(reg_pred, residuals)
```

***Performance metrics***

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred)) #Cost Function
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
#adjusted R square
1 - (1-score)*(len(y_test)-1) / (len(y_test) - x_test.shape[1] - 1)
```
